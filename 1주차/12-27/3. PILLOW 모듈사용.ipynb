## [2] PILLOW 모듈 실습

```python
#지정한 폴더내의 파일리스트보기
# os.listdir과 비교해서 폴더경로까지 보여줌
# .은 현재디렉토라
# ..은 전디렉토리

from glob import glob #glob 폴더에서 glob.py실행
from PIL import Image
from PIL import ImageFilter # 컨투어 필터링 선언


fileList=glob('./img/*.jpg')

file=fileList[2]

img=Image.open(file)
img=img.rotate(90)
img

img.filter(ImageFilter.CONTOUR) # 컨투어 필터링
```

```python
########################################
## 'K-mooc 이미지 처리 pillow.pdf 의 4페이지 변형
##  ---  이미지 폴더위치 추가 및 예외처리
## 이미지 열기 
## open 파일 없을시 예외처리
########################################
from PIL import Image
try:
    baseFolder='./img/'
    file= baseFolder + "flower01.jpg"
    im = Image.open(file)
    
    file=baseFolder+"flower05.jpg"
    
    im01 = Image.open(file)
    print(im.format, im.size, im.mode)
    print(im01.format, im01.size, im01.mode)
    
# IOError: 없는 파일을 열었을 때 어떻게 처리하는지 알아보기    
except IOError as err:  
    print("%s unable to load image"%file)
    
```

```python
########################################
## 이미지 열기
## 'imgList.csv' 파일에 있는 이미지 읽기 자료 리스트 읽기
## open 파일 없을시 예외처리
########################################
def imgOpen(file):
    from PIL import Image
    try:
        baseFolder='./img/'
        file= baseFolder + file
        im = Image.open(file)
        print(im.format, im.size, im.mode)
                
    # IOError: 없는 파일을 열었을 때 어떻게 처리하는지 알아보기    
    except IOError as err:  
        print("%s unable to load image"%file)
    
def prn():
    print('-'*50)
    
import csv
with open('imgList.csv', newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter='\t')
    for row in csvreader:
            prn()
            print(row)
            
            imgOpen(row[0])

```
