```python
import numpy as np
import matplotlib.pyplot as plt
def fileOpen(fileList):
    xData,yData=[],[]
    for file in fileList:
        img=Image.open(file).resize((200,300))
        img= np.array(img)
        xData.append(img)
        yData.append('flower')
    return xData,yData

def imgPltShow(num=0,**kwargs):
    for xD,yD in zip(kwargs['x'],kwargs['y']):
        plt.imshow(xD+num)
        plt.title(yD)
        plt.show()
```

```python
from glob import glob
fileList=glob('C:/Users/bluecom011/Desktop/2주차/img/*.jpg')
img,label=fileOpen(fileList)
imgPltShow(x=img,y=label)
```
