```python
def mysum(a):
    sum=0
    for i in a:
        sum+=i
    return sum
a=[1,2,3,4] # 원래는 리스트형이여야한다
print(mysum(a))
```

```python
def mysum(*a): # *을 함으로써 여러가지를 받을수 있다
    sum=0
    for i in a:
        sum+=i
    return sum
print(mysum(1,2,3,4))
```

```python
def mysum(*a):
    total_sum = 0
    for i in a:
        total_sum += i
    return total_sum

# Get input from the user
input_values = input("Enter space-separated numbers: ")

# Convert input values to a list of integers
input_list = [int(x) for x in input_values.split()]

# Call the function with the input list
result = mysum(*input_list)

# Print the result
print("Sum:", result)
```

```python
#함수화 시키기
import matplotlib.pyplot as plt
import numpy as np

def myImgShow(**kwargs):
    img = kwargs.get('img', []) #기본값은 빈 리스트
    label = kwargs.get('label', '')#기본값은 빈 문자열입니다.

    plt.imshow(img) #이미지를 플로팅
    plt.title(label) #플롯에 라벨을 설정
    plt.show() #플롯을 화면에 표시

    shape = np.shape(img) #이미지의 형태를 계산하고 반환
    return shape
```

```python
import matplotlib.pyplot as plt
import numpy as np
# 이미지 데이터
img = [
    [
        [[255, 0, 0], [0, 255, 0], [0, 0, 0], [255, 255, 0]],
        [[25, 0, 0], [0, 255, 255], [0, 0, 255], [255, 20, 0]]
    ],
    [
        [[255, 0, 0], [0, 255, 0], [0, 0, 255], [255, 255, 0]],
        [[25, 0, 0], [0, 255, 255], [0, 0, 0], [255, 255, 255]]
    ]
]
# 이미지 라벨
labels = ['dog', 'cat']

for x, y in zip(img, labels):
    plt.imshow(np.array(x)) # NumPy 배열로 변환 후 이미지 출력
    plt.title(y)
    plt.show()
```
