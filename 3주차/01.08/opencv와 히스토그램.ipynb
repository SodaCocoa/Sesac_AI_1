```python
import cv2
import sys
import matplotlib.pyplot as plt

src = cv2.imread(r'C:\Users\bluecom011\Desktop\01.08\lena.png', cv2.IMREAD_GRAYSCALE)

if src is None:
    print('Image load failed!')
    sys.exit()

# Calculate histogram
hist = cv2.calcHist([src], [0], None, [256], [0, 256])

# Display the image
cv2.imshow('src', src)
cv2.waitKey(0)  # Wait indefinitely until a key is pressed
cv2.destroyAllWindows()  # Close the OpenCV window

# Plot the histogram
plt.plot(hist)
plt.title('Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import cv2
import sys
import matplotlib.pyplot as plt

src = cv2.imread(r'C:\Users\bluecom011\Desktop\01.08\lena.png', cv2.IMREAD_COLOR)

if src is None:
    print('Image load failed!')
    sys.exit()

# Split the BGR channels
if src is None:
    print('Image load failed!')
    sys.exit()

colors = ['b', 'g', 'r']
bgr_planes = cv2.split(src)

for (p, c) in zip(bgr_planes, colors):
    hist = cv2.calcHist([p], [0], None, [256], [0, 256])
    plt.plot(hist, color=c)

cv2.imshow('src', src)
cv2.waitKey(1)

plt.show()
```
