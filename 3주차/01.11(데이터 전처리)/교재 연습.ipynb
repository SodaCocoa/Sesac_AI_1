```python
#1번
# CSV 파일 경로 설정
file_path = r'C:\Users\bluecom011\Desktop\3주차\01.11(데이터 전처리)\vaccine.csv'

# CSV 파일 읽기 (인코딩 지정)
df_vaccine = pd.read_csv(file_path, encoding='cp949')

# DataFrame 확인
print(df_vaccine)
```

```python
#2번 day1sum,day1rate,day2sum,day2rate 열 삭제

# 삭제할 열 목록
columns_to_drop = ['day1sum', 'day1rate', 'day2sum', 'day2rate']

# 열 삭제
df_vaccine_del = df_vaccine.drop(columns=columns_to_drop, axis=1)

# 수정된 DataFrame 출력
print(df_vaccine_del)

```

```python
#3번 기초통계량 확인
df_vaccine_del.describe()
```

```python
# CSV 파일 읽기 (인코딩 지정)
df_vaccine = pd.read_csv(file_path, encoding='cp949')

# 'date' 열을 날짜 형식으로 변환
df_vaccine['date'] = pd.to_datetime(df_vaccine['date'])

# 'date' 열에서 월 정보 추출하여 'month' 열 추가
df_vaccine['month'] = df_vaccine['date'].dt.month



# 'month' 열을 기준으로 평균 계산
monthly_mean = df_vaccine.groupby('month').mean()

# 열 삭제
monthly_mean1 = monthly_mean.drop(columns=columns_to_drop, axis=1)
# 월별 평균 출력
monthly_mean1.head()
```
