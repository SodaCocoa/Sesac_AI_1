```python
import requests

def search_naver_shopping(query, maker, max_price, client_id, client_secret):
    # 네이버 API 엔드포인트 및 파라미터 설정
    api_url = 'https://openapi.naver.com/v1/search/shop'
    headers = {
        'X-Naver-Client-Id': client_id,
        'X-Naver-Client-Secret': client_secret
    }
    params = {
        'query': query,
        'display': 100,  # 필요에 따라 페이지당 아이템 수 조절
        'sort': 'sim',
        'filter': 'all',
        'maker': maker,
        'lprice': max_price  # 0부터 max_price - 1까지의 가격 검색
    }

    # 네이버 쇼핑에서 검색 결과 가져오기
    response = requests.get(api_url, headers=headers, params=params)
    result = response.json()

    # JSON 형식으로 출력
    return result

# 네이버 API 클라이언트 ID와 클라이언트 시크릿 입력
your_client_id = ''
your_client_secret = ''

# 검색할 쿼리 및 조건 입력
search_query = '노트북'
maker = '삼성전자'
max_price = '1000000'  # 1,000,000원 이하

# 네이버 쇼핑 검색 함수 호출
result = search_naver_shopping(search_query, maker, max_price, your_client_id, your_client_secret)

# JSON 결과 출력
print(result)
```

```python
import requests
import json

# 네이버 API 관련 정보
naver_api_client_id = ""  # 발급받은 API 클라이언트 ID
naver_api_client_secret = ""  # 발급받은 API 클라이언트 Secret

# 검색할 키워드와 기간 설정
keyword = '노트북'
maker = '삼성전자'
max_price = '1000000'

# API 요청을 위한 함수 정의
def naver_book_search(query, maker, max_price, display=100):
    url = 'https://openapi.naver.com/v1/search/shop.json'
    headers = {
        'X-Naver-Client-Id': naver_api_client_id,
        'X-Naver-Client-Secret': naver_api_client_secret
    }
    result_list = []
    start = 1
    while True:
        params = {
            'query': query,
            'display': display,
            'start': start,
            'sort': 'sim',
            'filter': 'all',
            'maker': maker,
            'lprice': max_price
        }
        response = requests.get(url, params=params, headers=headers)
        result_json = response.json()
        if 'items' in result_json:
            # 필터링
            filtered_results = [
                item for item in result_json['items'] 
                if item.get('maker', '') == maker and int(item.get('lprice', 0)) <= int(max_price)
            ]
            result_list.extend(filtered_results)
            # 다음 페이지가 없으면 반복 종료
            if len(result_json['items']) < display:
                break
            # 다음 페이지의 시작 위치 설정
            start += display
        else:
            break
        return {'laptop': result_list}
    return result_list

# API 요청 및 결과 수신
search_results = naver_book_search(keyword, maker, max_price)

# 결과 출력 (JSON 형태)
print(json.dumps(search_results, ensure_ascii=False, indent=2))

output_file_path = 'naver_shop.json'
with open(output_file_path, 'w', encoding='utf-8') as json_file:
    json.dump(search_results, json_file, ensure_ascii=False, indent=2)

# Load JSON data from the file into a dictionary
with open(output_file_path, 'r', encoding='utf-8') as json_file:
    loaded_data = json.load(json_file)
```

```python
#가장 최저가 찾기
import requests

def search_naver_shop(query, client_id, client_secret):
    # 네이버 API 엔드포인트 및 파라미터 설정
    api_url = 'https://openapi.naver.com/v1/search/shop'
    headers = {
        'X-Naver-Client-Id': client_id,
        'X-Naver-Client-Secret': client_secret
    }
    params = {'query': query}

    # 네이버 쇼핑에서 검색 결과 가져오기
    response = requests.get(api_url, headers=headers, params=params)
    result = response.json()

    # JSON 형식으로 출력
    return result

# 네이버 API 클라이언트 ID와 클라이언트 시크릿 입력
your_client_id = ''
your_client_secret = ''

# 검색할 쿼리 입력
search_query = 'LG스마트TV'

# 네이버 쇼핑 검색 함수 호출
result = search_naver_shop(search_query, your_client_id, your_client_secret)

# JSON 결과 출력
# result['items'] is a list of items in the search result
items = result['items']

# Initialize variables to store the minimum price and corresponding link
min_price = float('inf')
min_price_link = None

# Iterate through the items to find the minimum 'lprice'
for item in items:
    title = item['title']
    lprice = int(item['lprice'])
    link = item['link']

    # Check if the current 'lprice' is lower than the minimum price
    if lprice < min_price:
        min_price = lprice
        min_price_link = link

# Print the link for the product with the lowest 'lprice'
print(f"Title: {title}")
print(f"Min Price: {min_price}")
print(f"Link: {min_price_link}")
```

```python
#각 제품별로 군집해서 가장 최저가찾고 링크 찾기
import requests
from collections import defaultdict

def search_naver_shop(query, client_id, client_secret):
    api_url = 'https://openapi.naver.com/v1/search/shop'
    headers = {
        'X-Naver-Client-Id': client_id,
        'X-Naver-Client-Secret': client_secret
    }
    params = {'query': query}

    response = requests.get(api_url, headers=headers, params=params)
    result = response.json()

    return result

your_client_id = ''
your_client_secret = ''
search_query = 'LG스마트TV'

result = search_naver_shop(search_query, your_client_id, your_client_secret)
items = result['items']

# 각 제품별 최소 가격과 링크를 저장할 딕셔너리 초기화
min_price_and_link_by_product = defaultdict(lambda: {'min_price': float('inf'), 'link': None})

# 각 항목의 'title'을 기준으로 군집화하고 최소 가격 및 링크를 갱신합니다.
for item in items:
    title = item['title']
    lprice = int(item['lprice'])
    link = item['link']

    # 현재 'title'의 최소 가격과 링크 갱신
    if lprice < min_price_and_link_by_product[title]['min_price']:
        min_price_and_link_by_product[title]['min_price'] = lprice
        min_price_and_link_by_product[title]['link'] = link

# 최소 가격과 링크를 가진 제품의 정보 출력
for title, data in min_price_and_link_by_product.items():
    print(f"제품명: {title}")
    print(f"최소 가격: {data['min_price']}")
    print(f"링크: {data['link']}")
    print("---")
```
