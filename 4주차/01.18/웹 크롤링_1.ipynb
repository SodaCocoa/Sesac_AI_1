```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

request = requests.get('https://music.bugs.co.kr/chart')
print(request) 
html = request.text
print(html)
soup = BeautifulSoup(html, 'html.parser')
```

```python
import pandas as pd

songs = soup.findAll('p', {'class' : 'title'})
artists = soup.findAll('p', {'class' : 'artist'})

# Create an empty list to store dictionaries
data = []

# Populate the list with dictionaries
for i in range(100):
    rank = i + 1
    song = songs[i].text.strip()
    artist = artists[i].text.strip()
    
    data.append({'Rank': rank, 'Song': song, 'Artist': artist})

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(data)
# Save DataFrame to a CSV file
df.to_csv('top_100_songs.csv', index=False, encoding='utf-8')
# Print the DataFrame
df.head(5)
```

```python
from bs4 import BeautifulSoup
import requests
from PIL import Image
from io import BytesIO

# 검색할 가수 입력
artist_name = input('가수의 이름을 입력하세요: ')

# 검색할 가수의 이미지 URL
url = f'https://music.bugs.co.kr/artist/{artist_name}'

# requests 패키지를 이용해 'url'의 html 문서 가져오기
response = requests.get(url)
response.encoding = 'utf-8'  # 한국어 웹페이지의 경우 UTF-8로 설정
html_text = response.text

# BeautifulSoup을 사용하여 파싱
soup = BeautifulSoup(html_text, 'html.parser')

# 가수의 이미지를 가져오기 위한 선택자를 찾아서 적용
# 예시로 가수의 이미지를 출력하고 저장하는 코드
image_element = soup.select_one('.artistImage img')
if image_element:
    image_url = image_element['src']
    image_response = requests.get(image_url)
    
    # 이미지 출력
    image = Image.open(BytesIO(image_response.content))
    image.show()

    # 이미지 저장
    image.save(f'{artist_name}_image.jpg')
    print(f'{artist_name}_image.jpg 파일로 저장되었습니다.')
else:
    print('이미지를 찾을 수 없습니다.')
```
