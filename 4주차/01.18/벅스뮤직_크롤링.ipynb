```python
# Step 1. 프로젝트에 필요한 패키지 불러오기
from bs4 import BeautifulSoup as bs
import requests

# Step 2. 검색할 키워드 입력
query = input('검색할 키워드를 입력하세요: ')

# Step 3. 입력받은 query가 포함된 url 주소(벅스 뮤직 검색 결과 페이지) 저장
url = 'https://music.bugs.co.kr/chart?wl_ref=M_contents_03_01'  # Corrected the URL

# Step 4. requests 패키지를 이용해 'url'의 html 문서 가져오기
response = requests.get(url)
print(response)

# Step 5. BeautifulSoup 패키지로 파싱 후, 'soup' 변수에 저장
soup = bs(response.text, 'html.parser')
print(soup)
```

```python
# Step 6. 아티스트 텍스트 추출
artist_elements = soup.select(".artist")  # Assuming the class name for artist information is 'artist'

for element in artist_elements:
    artist_text = element.get_text(strip=True)
    print(artist_text)
```

```python
# Step 7. 하이퍼링크 추출
for element in artist_elements:
    anchor_tag = element.find('a')  # Assuming the hyperlink is within an <a> tag
    if anchor_tag:
        href = anchor_tag.attrs.get('href')
        if href:
            print(href)
```

# 노래 제목, 아티스트, 하이퍼링크 정보 추출해서 csv로 저장하기

```python

import pandas as pd

# 노래 제목, 아티스트, 하이퍼링크 정보 추출
songs = soup.findAll('p', {'class' : 'title'})
artists = soup.findAll('p', {'class' : 'artist'})
thumbnail = soup.findAll('p', {'class' : 'thumbnail'})
# 빈 리스트를 생성하여 사전 형태의 데이터를 담습니다.
data = []

# 리스트를 딕셔너리로 채웁니다.
for i in range(100):
    rank = i + 1
    song_element = songs[i]
    artist_element = artists[i]
    thumbnail_element = thumbnail[i]
    # 노래와 아티스트 정보 추출
    song = song_element.text.strip()
    artist = artist_element.text.strip()

    # 'a' 태그 안의 'href' 링크를 가져오기 ('title' 클래스를 갖는 'p' 태그 안에 'a' 태그가 있다고 가정)
    link = song_element.find('a')['href']

    # 아티스트 요소에서 <a> 태그를 찾아 하이퍼링크 추출
    anchor_tag = artist_element.find('a')
    artist_link = anchor_tag.attrs.get('href') if anchor_tag else None

    # 데이터를 딕셔너리에 추가
    data.append({'순위': rank, '노래': song, '아티스트': artist, '앨범링크': link, '아티스트링크': artist_link})

# 데이터프레임을 딕셔너리 리스트로부터 생성합니다.
df = pd.DataFrame(data)

# 데이터프레임을 CSV 파일로 저장합니다.
df.to_csv('상위_100_노래.csv', index=False, encoding='utf-8')

# 데이터프레임을 출력합니다.
df.head(5)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import os

# 노래 정보 및 아티스트, 이미지 URL 추출
songs = soup.select('.title')
artists = soup.select('.artist')
image_urls = [img['src'] for img in soup.select('.track .thumbnail img')]

# 빈 리스트를 생성하여 사전 형태의 데이터를 담습니다.
data = []

# 이미지를 저장할 폴더 생성
image_folder = 'images'
os.makedirs(image_folder, exist_ok=True)

# 데이터를 딕셔너리로 채우고 이미지 다운로드 및 저장
for i, (song, artist, image_url) in enumerate(zip(songs, artists, image_urls), 1):
    song_text = song.get_text(strip=True)
    artist_text = artist.get_text(strip=True)
    image_filename = os.path.join(image_folder, f'{i:03d}_{song_text}_{artist_text}.jpg')

    if image_url:
        image_response = requests.get(image_url)
        with open(image_filename, 'wb') as img_file:
            img_file.write(image_response.content)

    data.append({'순위': i, '노래': song_text, '아티스트': artist_text, '이미지_파일명': image_filename})

# 하이퍼링크 추출 및 데이터에 추가
for i, element in enumerate(artist_elements, 1):
    anchor_tag = element.find('a')
    href = anchor_tag.attrs.get('href') if anchor_tag else ''
    data[i - 1]['노래_URL'] = href

# 데이터프레임 생성
df = pd.DataFrame(data)

# 데이터프레임을 CSV 파일로 저장
df.to_csv('상위_100_노래.csv', index=False, encoding='utf-8')

# 데이터프레임 출력
df.head(5)
```
