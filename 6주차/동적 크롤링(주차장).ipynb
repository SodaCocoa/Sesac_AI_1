```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys # 키보드 키사용
import csv
# 처음보는 작업임. driver.find_element_by_id를
# driver.find_element(By.ID )로 작업할수 있음.  이 노트차으이 19,20번 확인
from selenium.webdriver.common.by import By   
import time
import json
url='https://parking.seoul.go.kr/'

driver=webdriver.Chrome() # 크롬 드라이브 본인폴더로 이동

driver.get(url)  
time.sleep(2)


all_dropdown = driver.find_element(By.CSS_SELECTOR, 'a.val[title="구 선택"]')
all_dropdown.click()
time.sleep(2)  # 드롭다운 옵션이 보이도록 기다림

ul_element = driver.find_element(By.CSS_SELECTOR, "ul.select_list")
for li in ul_element.find_elements(By.TAG_NAME, "li"):
    if li.text == "도봉구":
        li.find_element(By.TAG_NAME, "a").click()
        break  # "도봉구"를 찾고 클릭하면 반복문 탈출

# "창동"을 클릭하기 전에 3초간 대기
time.sleep(1)

all_dropdown = driver.find_element(By.CSS_SELECTOR, 'a.val[title="동 선택"]')
all_dropdown.click()
time.sleep(1)  # 드롭다운 옵션이 보이도록 기다림

dong_options = driver.find_elements(By.CSS_SELECTOR, 'ul.select_list li a')
for option in dong_options:
    if option.text == "창동":
        option.click()
        break  # "창동"을 찾아 클릭하면 반복문 탈출


# 테마 검색 탭 선택
theme_tab = driver.find_element(By.XPATH, "//a[contains(text(), '테마검색')]")
theme_tab.click()
time.sleep(2)
# 장애인 주차 테마 선택
theme_link = driver.find_element(By.XPATH, "//a[contains(text(), '장애인주차')]")
theme_link.click()
#확인버튼누르기
time.sleep(1)
search_button = driver.find_element(By.ID, "parking_search")
search_button.click()
time.sleep(1)
# 스크린샷 저장
screenshot_path = r"C:\Users\bluecom011\Desktop\Sesac_AI\공모전/screenshot.png"  # 저장할 파일의 경로와 이름
driver.save_screenshot(screenshot_path)

# # 작업 완료 후 드라이버 종료 (필요한 경우)
# driver.quit()

print(f"스크린샷이 {screenshot_path}에 저장되었습니다.")
#####################################################################################################################################
# 웹페이지에서 정보를 추출하는 부분
# "pk_search_view" id를 가진 ul 요소에서 모든 li 요소를 찾음
results_list = driver.find_element(By.ID, "pk_search_view")
parking_lots = []
address_element = "#pk_search_view > ul:nth-child(1) > li > strong"
links_selector = "#pk_search_view > ul:nth-child(1) > li > span:nth-child(3)"
# 각 li 요소에서 필요한 정보(예: 주차장 이름, 주소 등)를 추출하여 리스트에 저장
for li in results_list.find_elements(By.TAG_NAME, "li"):
    parking_name_element = li.find_element(By.CSS_SELECTOR, "strong.tit > a")  # 주차장 이름을 포함하는 <a> 태그
    parking_name = parking_name_element.text.strip()
    
    # Corrected to use relative XPath
    # address_element = li.find_element(By.XPATH, ".//span[contains(text(), '도로주소')]")  # 주차장 도로주소를 포함하는 <span> 태그
    address = address_element.text.replace('도로주소 : ', '').strip()

    parking_lot = {
        "name": parking_name,
        "address": address,
    }
    parking_lots.append(parking_lot)

# 추출한 데이터를 JSON 파일로 저장
json_file_path = r"C:\Users\bluecom011\Desktop\Sesac_AI\공모전/parking_lots.json"  # JSON 파일 저장 경로 (경로는 환경에 맞게 수정하세요)
with open(json_file_path, 'w', encoding='utf-8') as file:
    json.dump(parking_lots, file, ensure_ascii=False, indent=4)

print(f"데이터가 {json_file_path}에 JSON 형식으로 저장되었습니다.")

# 작업 완료 후 드라이버 종료
driver.quit()
```
