```python
import pandas as pd
import numpy as np
import folium
import geopy
excel_file_path = r'C:\Users\bluecom011\Desktop\공모전\주차구역\서울특별시 종로,은평,용산,성북,광진_장애인주차구역_20220615.xlsx'

# 엑셀 파일 읽기 (엔진으로 'openpyxl'을 명시)
df = pd.read_excel(excel_file_path, engine='openpyxl')
```

```python
# 위도
latitude = 37.5665
# 경도
longitude = 126.9780
```

```python
from folium.plugins import MarkerCluster


m = folium.Map(
    location=[latitude, longitude],
    zoom_start=15
)

coords = df[['Latitude', 'Longitude']]


marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(coords['Latitude'], coords['Longitude']):
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)
m
```

```python
import requests
import json

# 서울 행정구역 json raw파일(githubcontent)
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
import pandas as pd
import numpy as np
import folium
import geopy
m = folium.Map(
    location=[latitude, longitude],
    zoom_start=15,
    tiles='cartodbpositron'
)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(m)

marker_cluster = MarkerCluster(disableClusteringAtZoom=5).add_to(m)

for lat, long in zip(coords['Latitude'], coords['Longitude']):
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)

m
# m.save('map_with_markers_and_heatmap.html')
```
