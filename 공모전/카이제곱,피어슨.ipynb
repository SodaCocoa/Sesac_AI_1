## 군집분석

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 데이터 불러오기 (불법주차 건수, 센서 입지 등의 변수 포함)
# 여기서는 예시로 임의의 데이터를 사용하겠습니다. 실제 데이터에 맞게 수정 필요
file_path = 'your_data.csv'
data = pd.read_csv(file_path)

# 필요한 변수 선택 (예: 불법주차 건수, 센서 입지 변수 등)
selected_features = ['불법주차건수', '센서입지1', '센서입지2', ...]

# 선택한 변수만으로 데이터프레임 생성
X = data[selected_features]

# 데이터 표준화 (Standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 군집분석 (K-means 알고리즘 사용)
kmeans = KMeans(n_clusters=3, random_state=42)  # 군집의 개수는 조절 가능
data['Cluster'] = kmeans.fit_predict(X_scaled)

# 군집 결과 확인
print(data[['불법주차건수', 'Cluster']])

# 군집별 특성 확인 (평균 값 등)
cluster_summary = data.groupby('Cluster').mean()
print(cluster_summary)

# 시각화 등을 통해 군집 결과를 살펴봅니다.
# 예시로 2차원 시각화 (실제 데이터에 따라 조정 필요)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=data['Cluster'], cmap='viridis')
plt.title('Cluster Analysis')
plt.xlabel('불법주차건수 (Standardized)')
plt.ylabel('센서입지1 (Standardized)')
plt.show()
```
