```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# 텍스트 전처리
# HTML 태그 제거 및 특수문자 제거는 이미 'str' 열에서 처리된 것으로 보임
# 추가적인 전처리가 필요한 경우 여기에 구현할 수 있음

# TF-IDF 벡터화
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df_result['str'])

# 벡터화된 데이터의 크기 확인
tfidf_matrix.shape
```

```python
import time

# 코드 실행 시작 시간 기록
start_time = time.time()

# 'str' 열을 사용해 텍스트 데이터를 가져옵니다.
text_content = df_result['str']

# TF-IDF 벡터화를 위한 TfidfVectorizer 객체를 생성합니다.
vectorizer = TfidfVectorizer(
    max_df=0.4,          # 문서에서 너무 자주 등장하는 단어는 무시합니다. (40% 이상)
    min_df=1,            # 최소한 1개의 문서에는 등장해야 합니다.
    lowercase=True,      # 모든 단어를 소문자로 변환합니다.
    use_idf=True,        # IDF를 사용하여 가중치를 적용합니다.
    norm=u'l2',          # 벡터의 크기를 조절하는 방법을 설정합니다. 여기서는 L2 정규화를 사용합니다.
    smooth_idf=True      # IDF 계산 시 분모가 0이 되는 경우를 방지하기 위해 1을 더합니다.
)

# TF-IDF 벡터화를 수행합니다.
tfidf_matrix_custom = vectorizer.fit_transform(text_content)

# 코드 실행 시간을 측정합니다.
elapsed_time = time.time() - start_time

elapsed_time
```

```python
from sklearn.cluster import MiniBatchKMeans

# 클러스터의 개수 설정
k = 200

# MiniBatchKMeans 모델을 생성하고 데이터에 맞춰 학습합니다.
kmeans = MiniBatchKMeans(n_clusters=k, random_state=0)
kmeans.fit(tfidf_matrix)

# 클러스터의 중심을 기준으로 각 클러스터 내에서 가장 중요한 단어들의 인덱스를 저장합니다.
centers = kmeans.cluster_centers_.argsort()[:, ::-1]

# TF-IDF 벡터화에서 사용된 단어들을 가져옵니다.
terms = vectorizer.get_feature_names_out()

# 각 클러스터에 속한 데이터 포인트가 속한 클러스터를 새로운 열 'cluster'에 저장합니다.
df_result['cluster'] = kmeans.predict(tfidf_matrix)

# 각 클러스터에 속한 데이터 포인트의 개수를 확인하여 상위 5개 클러스터를 출력합니다.
cluster_counts = df_result['cluster'].value_counts().head()

cluster_counts
```

```python
# find_similar 함수 정의 (판례 데이터에 맞게 수정)
def find_similar(tfidf_matrix, index, top_n=5):
    # TF-IDF 벡터화된 행렬에서 주어진 인덱스의 판례와 다른 모든 판례 간의 코사인 유사도를 계산합니다.
    cosine_similarities = linear_kernel(tfidf_matrix[index:index+1], tfidf_matrix).flatten()
    # 유사도를 기준으로 내림차순으로 정렬하여 인덱스를 가져옵니다. 자기 자신은 제외합니다.
    related_docs_indices = [i for i in cosine_similarities.argsort()[::-1] if i != index]
    # 상위 top_n개의 유사한 판례 인덱스를 반환합니다.
    return [index for index in related_docs_indices][:top_n]

# 가정: 사용자의 질문 또는 텍스트는 이미 벡터화되었고, 해당 벡터는 tfidf_matrix에 추가되었다고 가정합니다.
# 이 예제에서는 tfidf_matrix의 첫 번째 문서(인덱스 0)와 가장 유사한 문서 5개를 찾습니다.
# 실제로는 사용자의 질문을 tfidf_matrix에 추가하고 그 인덱스를 사용해야 합니다.

# 사용자 질문에 가장 유사한 판례 찾기 (예제로 첫 번째 문서를 사용)
top_n_similar_indices = find_similar(tfidf_matrix, 0, top_n=5)

# 결과 출력
top_n_similar_indices
```

```python
import networkx as nx
import time

# 네트워크 그래프 생성
G = nx.Graph(label="Case")

start_time = time.time()

# 데이터프레임을 반복하면서 각 행을 처리합니다.
for i, rowi in df_result.iterrows():
    # 매 100번째 반복마다 경과 시간을 출력합니다.
    if i % 100 == 0:
        print(" iter {} -- {} seconds --".format(i, time.time() - start_time))
    
    # 판례 노드 추가
    G.add_node(i, label="Case", text=rowi['str'])
    
    # 유사한 판례 찾기
    indices = find_similar(tfidf_matrix, i, top_n=5)
    snode = "Sim(Case {})".format(i)
    G.add_node(snode, label="SIMILAR")
    G.add_edge(i, snode, label="SIMILARITY")
    
    # 유사한 판례들과의 관계 추가
    for index in indices:
        G.add_edge(snode, index, label="SIMILARITY")

# 처리가 완료되었음을 출력합니다.
print(" finish -- {} seconds --".format(time.time() - start_time))
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

def visualize_subgraph(G, root_case):
    """특정 형사 사건과 연결된 사건을 포함하는 부분 그래프를 시각화합니다."""
    # 특정 형사 사건과 연결된 사건을 포함하는 부분 그래프를 생성
    sub_graph_nodes = [root_case] + list(G.neighbors(root_case))
    subgraph = G.subgraph(sub_graph_nodes)

    # 부분 그래프 시각화
    plt.figure(figsize=(8, 6))
    colors = ['red' if node == root_case else 'blue' for node in subgraph.nodes()]
    nx.draw(subgraph, with_labels=True, node_color=colors, font_weight='bold', node_size=700)
    plt.title(f"사건 {root_case} 및 관련 사건에 대한 부분 그래프")
    plt.show()

# 네트워크 그래프 G 생성 예시
G = nx.Graph()
for i in range(5):  # 5개의 형사 사건을 가정
    G.add_node(i, label="형사 사건", text=f"사건 {i}")
G.add_edge(0, 1)
G.add_edge(0, 2)
G.add_edge(3, 4)

# 사용자로부터 형사 사건의 식별자 입력 받기
user_input = input("관심 있는 형사 사건의 식별자를 입력하세요 (예: 0, 1, 2, ...): ")

# 입력값 검증 및 부분 그래프 시각화
try:
    root_case = int(user_input)
    if root_case in G.nodes():
        visualize_subgraph(G, root_case)
    else:
        print("네트워크에 지정된 사건이 존재하지 않습니다.")
except ValueError:
    print("유효한 정수 식별자를 입력하세요.")
```

```python
import math
import numpy as np
import pandas as pd

def get_recommendation(root):
    # 공통 이웃을 저장하는 딕셔너리 생성
    commons_dict = {}
    # 주어진 판례의 이웃을 반복
    for e in G.neighbors(root):
        # 이웃의 이웃을 다시 반복
        for e2 in G.neighbors(e):
            # 자기 자신은 제외
            if e2 == root:
                continue
            # 이웃이 판례인 경우
            if G.nodes[e2]['label'] == "Case":
                commons = commons_dict.get(e2)
                # 공통 이웃 딕셔너리 업데이트
                if commons is None:
                    commons_dict[e2] = [e]
                else:
                    commons.append(e)
                    commons_dict[e2] = commons
    # 추천할 판례와 가중치를 저장할 리스트 생성
    cases = []
    weight = []
    # 공통 이웃 딕셔너리를 순회하면서 가중치 계산
    for key, values in commons_dict.items():
        w = 0.0
        # 이웃의 가중치를 계산하여 합산
        for e in values:
            w += 1 / math.log(G.degree(e))
        # 판례와 가중치를 각각의 리스트에 추가
        cases.append(key)
        weight.append(w)
    
    # 결과를 Series로 변환하여 가중치에 따라 정렬
    result = pd.Series(data=np.array(weight), index=cases)
    result.sort_values(inplace=True, ascending=False)        
    return result

# 예제 사용: 특정 판례에 대한 추천 받기
# 이 예제를 실행하기 전에 'G' 네트워크 그래프가 생성되어 있어야 하며,
# 'root' 변수는 추천을 원하는 판례의 식별자로 설정해야 합니다.
root = 0  # 예제로 첫 번째 판례를 사용
recommendations = get_recommendation(root)

# 추천 결과 출력
print(recommendations)
```

```python
# 판례 데이터에 대한 추천 목록을 가져오는 예시
# 판례 인덱스를 사용하여 추천을 받습니다. 실제 인덱스 값을 알아야 합니다.
# 여기서는 예제로 인덱스 0, 1, 2, 3에 해당하는 판례를 사용합니다.

# 결과를 저장할 변수를 초기화합니다.
results = []

# 판례 데이터의 예시 인덱스
example_cases = [0, 1, 2, 3]  # 실제 판례 인덱스에 해당하는 값을 사용해야 합니다.

# 각 판례에 대한 추천 목록을 가져옵니다.
for index in example_cases:
    result = get_recommendation(index)
    results.append(result)

# 각 판례에 대한 추천 목록을 출력합니다.
titles = ["매매", "회사", "판단", "심판"]  # 예시로 사용된 판례의 제목 또는 식별자
for i, result in enumerate(results):
    print("*" * 40 + f"\n Recommendation for '{titles[i]}'\n" + "*" * 40)
    print(result.head())
```
